```python
import re
import pandas as pd
```

```python
# Opens the WhatsApp chat text file in read mode with UTF-8 encoding to handle any special characters
f = open('WhatsApp Chat with ITMVU-8TH-B.TECH-CSE(2021-25).txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
# Defining a regular expression pattern to match the date and time format used in WhatsApp messages
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
# Splitting the chat data using the defined date-time pattern to extract the actual messages
messages = re.split(pattern,data)[1:]
messages
```

```python
# Extracting all date-time occurrences from the chat data using the regular expression pattern
dates = re.findall(pattern,data)
dates
```

```python
# Creating a DataFrame with messages and corresponding dates and converting string dates into datetime objects
df = pd.DataFrame({'user_message':messages,'message_date':dates})
df['message_date'] = pd.to_datetime(df['message_date'],format='%m/%d/%y, %H:%M - ')
df.rename(columns={'message_date':'date'},inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
# separate users and messages
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'],inplace=True)
df.head()
```

```python
# Extracting various datetime features (year, month, day, hour, minute) from the 'date' column
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
period = []
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period'] = period
```
